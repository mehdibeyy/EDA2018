### Analyse Uni_Variée

```python
#Création du Lineplot
sns.lineplot(x='state', y='backers', data=df)
plt.title('La réussite selon les backers')
```

On remarque plus le nombre de Backers augmente ,plus on assure la reussite du projet

```python
sns.lineplot(x='state', y='pledged', data=df)
plt.title('La réussite selon les pledged')
```

plus on engage plus d'argent ,plus la réussite du projet augmente

```python
sns.lineplot(x='state', y='usd_pledged_real', data=df)
plt.title('La réussite selon les usd_pledged_real')
```

La réussite d'un projet augmente quand l'engagement réel(usd_pledged_real) augmente
